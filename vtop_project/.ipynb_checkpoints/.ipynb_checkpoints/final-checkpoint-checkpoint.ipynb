```python
def myfunc(AddTheorySlot="",RemoveTheorySlot="",AddLabSlot="",RemoveLabSlot=""):
    cell_width = .5
    cell_height = 2
    fig,ax = plt.subplots()
    for j in range(1,12,2):
        for i in range(0,9):
            ax.add_patch(plt.Rectangle((i*cell_width, j*cell_height), cell_width, cell_height, fill=True, facecolor='lightyellow', edgecolor='black'))
    for j in range(0,12,2):
        for i in range(0,11):
            ax.add_patch(plt.Rectangle((i*cell_width, j*cell_height), cell_width, cell_height, fill=True, facecolor='beige', edgecolor='black'))
    for i in range(12):
        ax.add_patch(plt.Rectangle((4*cell_width, i*cell_height), cell_width, cell_height, fill=True, facecolor='lightblue', edgecolor='black'))
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            ax.text(j*cell_width+cell_width/2, i*cell_height+cell_height/2, grid[i][j], ha='center', va='center', color='black')
    ax.text(-1.5*cell_width+cell_width/2, 11*cell_height+cell_height/2, "Monday", ha='center', va='center', color='black',)
    ax.text(-1.5*cell_width+cell_width/2, 9*cell_height+cell_height/2, "Tuesday", ha='center', va='center', color='black',)
    ax.text(-1.5*cell_width+cell_width/2, 7*cell_height+cell_height/2, "Wednesday", ha='center', va='center', color='black',)
    ax.text(-1.5*cell_width+cell_width/2, 5*cell_height+cell_height/2, "Thursday", ha='center', va='center', color='black',)
    ax.text(-1.5*cell_width+cell_width/2, 3*cell_height+cell_height/2, "Friday", ha='center', va='center', color='black',)
    ax.text(-1.5*cell_width+cell_width/2, 1*cell_height+cell_height/2, "Saturday", ha='center', va='center', color='black',)
    ax.text(0*cell_width+cell_width/2, 12*cell_height+cell_height/2, "9-10", ha='center', va='center', color='black',)
    ax.text(1*cell_width+cell_width/2, 12*cell_height+cell_height/2, "10-11", ha='center', va='center', color='black',)
    ax.text(2*cell_width+cell_width/2, 12*cell_height+cell_height/2, "11-12", ha='center', va='center', color='black',)
    ax.text(3*cell_width+cell_width/2, 12*cell_height+cell_height/2, "12-1", ha='center', va='center', color='black',)
    ax.text(5*cell_width+cell_width/2, 12*cell_height+cell_height/2, "2-3", ha='center', va='center', color='black',)
    ax.text(6*cell_width+cell_width/2, 12*cell_height+cell_height/2, "3-4", ha='center', va='center', color='black',)
    ax.text(7*cell_width+cell_width/2, 12*cell_height+cell_height/2, "4-5", ha='center', va='center', color='black',)
    ax.text(8*cell_width+cell_width/2, 12*cell_height+cell_height/2, "5-6", ha='center', va='center', color='black',)
    ax.text(9*cell_width+cell_width/2, 12*cell_height+cell_height/2, "6-7", ha='center', va='center', color='black',)
    ax.text(10*cell_width+cell_width/2, 12*cell_height+cell_height/2, "7-8", ha='center', va='center', color='black',)
    ax.set_xticks(np.arange(0, 11*cell_width, cell_width))
    ax.set_yticks(np.arange(0, 12*cell_height, cell_height))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.grid(which='major', color='black', linestyle='-', linewidth=1)
    ax.set_xlim([0, 11*cell_width])
    ax.set_ylim([0, 12*cell_height])
    if(AddTheorySlot!=""):
        selectSet.add(AddTheorySlot)
    if(RemoveTheorySlot!=""):
        if(RemoveTheorySlot in selectSet):
            selectSet.remove(RemoveTheorySlot)
    if(AddLabSlot!=""):
        selectSet.add(AddLabSlot)
    if(RemoveLabSlot!=""):
        if(RemoveLabSlot in selectSet):
            selectSet.remove(RemoveLabSlot)
    for slot in selectSet:
        if(slot==""):
            continue
        for tup in slotcells[slot]:
            a = tup[0]
            b=tup[1]
            ax.add_patch(plt.Rectangle((b*cell_width, a*cell_height), cell_width, cell_height, fill=True, facecolor='lightgreen', edgecolor='black'))
        for tup in clashcells[slot]:
            a = tup[0]
            b=tup[1]
            ax.add_patch(plt.Rectangle((b*cell_width, a*cell_height), cell_width, cell_height, fill=True, facecolor='pink', edgecolor='black'))
    

    plt.show()
```
